```python
from scipy.stats import chisquare
f_exp = [ 0.2 ,0.1, 0.4 ,0.3]
f_obs = [62 ,19 ,87,76]
chisquare(f_obs, f_exp=[p * 244 for p in f_exp])
```

```python
from scipy.stats import ttest_1samp

tips = sns.load_dataset("tips")

ttest_1samp(tips['tip'], popmean=3.5)
```

```python
# la correlation entre size et day 
import pandas as pd
table = pd.crosstab(df['day'], df['size'])
print(table)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
chi2, p_val, dof, expected = chi2_contingency(table)

print(f"Statistique Khi-deux : {chi2:.4f}")
print(f"P-value : {p_val:.4f}")
```
